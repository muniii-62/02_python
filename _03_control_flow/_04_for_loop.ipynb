# for 반복문

- `for` 반복문은 iterable 객체의 요소를 하나씩 꺼내 실행한다.
- `str`, `list`, `tuple`, `dict`, `set`, `range` 등이 대표적인 iterable 객체이다.
- 반복 횟수가 정해져 있거나 컬렉션의 모든 요소를 순회할 때 자주 사용한다.

## 1. 문자열, 리스트, 튜플 순회

문자열은 문자 단위로, 리스트와 튜플은 요소 단위로 순회한다.(시퀀스 타입)

```python
text : str = 'hello world'
for ch in text :
    print(ch)

lst : list[int] = [10,20,30]
for num in lst :
    print(num)

tpl : tuple[int,int,int] = (30, 35, 40)
for num in tpl :
    print(num)
```

## 2. dict와 set 순회

`dict`를 그대로 순회하면 key가 나온다. `items()`를 사용하면 key와 value를 함께 꺼낼 수 있다.

`set`은 순서가 없는 자료형이므로 출력 순서는 실행 환경에 따라 달라질 수 있다.

```python
from typing import Any

#type hint - dict[str,Any]에서Any
# - 어떤 타입이든 작성 가능하다는 것을 나타내는 표현(클래스)


user_info : dict[str,Any] = {
    'name' : '홍길동',
    'age' : 23,
    'hobby' : ['독서','유튜브']
}

#dict 순회 -> key 순회, value 순회, item 순
print('--- dick key 순회---')
# for A in B
# A : 객체에서 꺼내와 저장할 변수
# B : 순회 가능한 객체(str, list, tuple, dict, set, ...)

for key in user_info :
    print(f'{key} : {user_info[key]}')


print('--- dict item 순회')
for key,value in user_info.items() :
    print(f'{key} : {value}')


print('---set 순회 : if 를 곁들인...---')
nums : set = {1,2,3,4,5,6,3,2,4,5,6,7,8,9,5,1,4,23,21,222,5,4,213,213}
result : list[int] = []
for i in nums :
    if i % 2 == 0 :
        result.append(i)
print(result)
```

## 3. range 객체

`range(start, end, step)`은 정수 범위를 표현하는 객체이다.

- `start`: 시작값
- `end`: 끝값, 포함되지 않음
- `step`: 증가 또는 감소 간격

```python
#rnage 객체 : 지정된 범위 내 정수를 step 만큼 증가하면서 값을 하나씩 반ㅇ환하는 객체

print("---range 객체 ---")
rng = range(1,5,1)
print(rng,type(rng))

for num in rng :
    print(num)

print('---range 객체를 이용한 정수 list 만들기---')
nums : list[int] = list(range(1,11,1))
print("nums:",nums)

print('---1부터 100까지 모든 정수 합 구하기---')
sum=0
for n in range(1,101,1) :
    sum+=n
print(sum)

print('---1부터 100까지 모든 홀수 합 구하기---')
sum1=0
for n in range(1,101,1) :
    if n % 2 != 0 :
        sum1+=n
print(sum1)
```

## 4. 중첩 반복문

반복문 안에 반복문을 넣으면 행과 열처럼 2차원 구조를 만들 수 있다. 구구단과 별찍기 같은 예제에 자주 사용한다.

```python
print('---중첩 반복문 예시---')
for a in range(1,4) :
    for b in range(1,3) :
        print(a,b)


```

```python
print('---2단 출력하기---')
dan : int = 2
for i in range(1,10) :
    print(f'{dan} x {i} = {dan*i}')
```

```python
print('---구구단 전체 출력(2단 ~9단)---')

for i in range(2,10) :
    for j in range(1,10) :
        print(f'{i} x {j} = {i*j}')
    print('-'*10)
```

```python
print('---range를 이용한 반복 횟수 설정')
print('🍗3행 5열 출력')
print('-'*14)
for row in range(3) :
    for col in range(5) :
        print('🍗', end='')
    print()
```

## 5. zip과 enumerate

- `zip()`은 여러 iterable을 같은 순서끼리 묶는다.
- `enumerate()`는 인덱스와 요소를 함께 꺼낸다.

```python
print('--- zip 객체 ---')
names : list[str] = ['a','b','c']
nums : list[int] = [10,20,30]

z = zip(names, nums)
# print(z,type(z))

for name, num in z :
    print(name,num)

print('---enumerate 객채---')
foods : list[str] = ['제육','피자','치킨','파스타']

for i,food in enumerate(foods) :
    print(f'{i}번 인덱스 : {food}')
```

## 6. 반복문으로 평균 구하기

반복문을 사용하면 리스트나 딕셔너리 리스트에 들어 있는 값을 누적해서 평균을 계산할 수 있다.

```python
print('---list에 담긴 값 평균 구하기---')
scores : list[int] = [23,42,56,8,100]
sum_score = 0
for score in scores :
    sum_score += score
avg = sum_score / len(scores)
print('평균:', avg)
print('평균:', sum(scores)/len(scores))
```

```python
print('---list[dict]에서 점수 평균 구하기')

students: list[dict[str,object]] = [
    {'name':'홍길동','score':100},
    {'name':'신사임당','score':77},
    {'name':'이순신','score':85},
]

students[2].get('score')

sums : int = 0
for std in students :
    sums += std.get('score')

avg= round(sums/len(students))
print('평균:',avg)
```

```python
# range 9 indexing

for index in range(len(students)) :
    print(students[index].get('score'))
```
