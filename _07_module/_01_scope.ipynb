# 유효범위 Scope

변수의 유효범위(scope)는 변수에 접근할 수 있는 범위를 의미한다.

- 전역 변수(global variable): 함수 블록 밖에서 선언되어 여러 위치에서 접근할 수 있는 변수
- 지역 변수(local variable): 함수 내부에서 선언되어 해당 함수 안에서 주로 사용하는 변수
- Python의 `for`, `if` 블록은 Java의 중괄호 블록처럼 별도의 지역 스코프를 만들지 않는다.


## 1. 전역 변수와 지역 변수

함수 밖에서 선언된 `a`는 전역 변수이고, 함수 안에서 선언된 `b`는 지역 변수이다.
`b`는 `foo()` 함수 내부에서만 사용할 수 있다.


```python
a = 100 # Global
def foo() :
    print('foo()에서 출력한 a:',a)

    b = 200 # foo() 함수 내에서만 사용 가능
    print('foo()에서 출력한 b :',b)

foo()
print('함수밖에서 출력한 a: ',a)
# print('함수밖에서 출력한 b: ',b) # 확인되지 않은 참조'b'
```

## 2. for, if 블록과 스코프

Python에서는 `for`, `if` 블록이 별도의 지역 스코프를 만들지 않는다.
따라서 블록 안에서 만들어진 이름을 블록 밖에서도 참조할 수 있다.


```python
for i in range(3) :
    print('for문 내 i : ',i)
    j=100

print('for문 종료 후 : ',i)
print(j)


if j == 100 :
    k = 200

print('k의 값은?',k)
```

## 3. 이름 탐색 우선순위

같은 이름의 변수가 여러 범위에 있을 때는 더 가까운 범위의 변수가 우선된다.
함수 안에서 같은 이름으로 대입하면 전역 변수를 바꾸는 것이 아니라, 지역 변수를 새로 만든다.


```python
x = 100

def bar() :
    x = 200
    print('bar() 내부 x: ', x)


print('bar() 외부 x : ', x) #bar에서한거는 보이지 않음
```

## 4. 전역 변수와 지역 변수가 함께 사용되는 경우

아래 예제에서 `koo()` 함수 안의 `glo`는 지역 변수이고, 함수 밖의 `glo`와는 다른 이름 공간에 있다.


```python
glo = 10
bal = 5

def koo(a:int) -> int :
    glo = a+ 10
    print('koo()함수 내 glo : 출력',glo) #지역 변수 glo
    b = a + glo + bal #지역a, 지역glo + 전역bal
    return b
koo(7)
print('koo()함수 밖 glo : 출력',glo) # 전역 변수 glo
print('koo(7)의 반환 값 : ',koo(7))
```

## 5. immutable, mutable 객체 전달

Python은 함수에 값을 전달할 때 객체에 대한 참조를 전달한다.
다만 immutable 객체는 내부 값을 직접 바꿀 수 없고, mutable 객체는 내부 값을 바꿀 수 있어 결과가 다르게 보인다.

- immutable: `int`, `float`, `bool`, `tuple`, `str`
- mutable: `list`, `dict`, `set`


```python
    def func1(a:int) :
        a = 200 #전달받은 a에 새 값을 대임(수정) ->새로운 지역변수 a에 200 대입
        print('func1()내부a',a,id(a))
        return a
    a = 100
    print('func1() 호출 전 a:',a,id(a))

    func1(a)
    print('func1() 호출 후 a:',a,id(a))
```

```python
def func2(b:list)-> None :
    b[0] = 999

b=[1,2,3]
print('func2() 호출 전 b:',b,id(b))

func2(b)
print('func2() 호출 후 b:',b,id(b))
```

숫자, 문자열 → 값 복사 → 함수 안에서 바꿔도 밖에 영향 없음
리스트, 딕셔너리 → 주소 전달 → 함수 안에서 바꾸면 밖에도 반영됨
