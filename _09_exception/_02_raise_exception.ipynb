```python
class AdultGame :
    def __init__(self) :
        pass
    def start(self):
        print('성인용 게임(고스톱)을 시작합니다.')

# if -else
age = int(input('나이 입력: '))
if age >= 19 :
    AdultGame().start()
else :
    print('돌아가')

#if-else 가 잘못된 것은 아니지만코드 흐름의 의도와맞지는않다.

```

```python
# 사용자 정의 예외 클래스
# 만드는 방법 : 기존 Exception , Error 클래스 중 아무거나 상속 받기
class UnderAgeError(Exception) :
    pass
```

```python
class AdultGame2 :
    def __init__(self,age) :
        if age < 19 :
            #BaseException : 모든 예외, 에러 클래스의 최상위 부모 클래수
            #raise BaseException(f'미성년자는 이용할 수 없습니다.:{age}')

            # 상황에 맞는 예외 강제 발생
            #파이썬이 제공해주는 알맞은 예외가없으면 사용자가 생성 가능
            raise UnderAgeError((f'미성년자는 이용할 수 없습니다.:{age}'))
    def start(self):
        print('성인용 게임(고스톱)을 실행합니다.')

try :
    age = int(input('나이 입력: '))
    AdultGame2(age).start()

except UnderAgeError as e:
    print(type(e).__name__)
    print(e)

print('종료')
```
