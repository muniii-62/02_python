# 추상클래스 Abstract Class
- 자식클래스에서 구현해야 할 추상메서드를 가진 부모클래스 (인터페이스)
-   - 추상 메서드 == 미완성 메서드 정의 부분 비어있음.
-
- **추상메서드는 자식클래스에서 반드시 구현(오버라이딩)해야할 의무를 가진다.**
- 추상클래스는 직접 객체화 할수 없다.

- abc.ABC클래스를 상속하면 추상클래스를 선언할 수 있다.(추상 클래스로 인스턴스 생성 불가)
    - ABC == Abstract Base Class
- @abstractmethod 데코레이터로 추상메소드를 선언할 수 있다.

```python
from abc import ABC,abstractmethod

#추상 클래스 정의
#ABC 상속 == 해당 클래스가 너무 추상적이거나 또는 해당 틀래스로 인스턴스 생성을 안하겠다 라는 의미
class Product(ABC):
    #class 아래는 부모부분을 만들어주는거라고 할 수 있씀
    def __init__(self,code:str,name:str,price:float):
        self.__code = code
        self.__name= name
        self.__price = price

    # 추상 메서드 == 미완성 메서드 == 정의 부분이 없음
    @abstractmethod
    def get_tax(self):
        pass # 정의 부분을 상속 받은 자식에게 넘김

    def get_price(self):
        return self.__price

    def __str__(self):
        return f'code={self.__code} name={self.__name} price={self.__price}'
```

## 여러 자식 클래스를 부모 타입으로 다루기

다형성을 이용하면 서로 다른 자식 클래스가 같은 부모 규격을 따르는 모습을 볼 수 있다.

```python
#추상 클래스로 인스턴스 생성하기  #오류 확인
#prod =Product("111","노트북",2000000)
```

```python
# 추상 클래스를 상속 받은 자식 클래스
class TV(Product):
    def __init__(self,code:str,name:str,price:float,size:int):
        super().__init__(code, name, price)
        self.size = size

    #상속받은 추상 메서드ㅡㄹ 재작성 == 오버라이딩
    def get_tax(self): #부가세를 10%로 책정해서 반환
        return self.get_price()*0.1

    def __str__(self):
        return f'{super().__str__()},size={self.size}'

```

```python
tv1 = TV("samsung-tv-1234", "SAMSUNG OLED TV", 5_000_000, 85)
print(tv1)
print(tv1.get_tax())

tv2 = TV("lg-tv-0001", "LG Flexible TV", 10_000_000, 100)
print(tv2)
print(tv2.get_tax())
```

```python
class Book(Product) :
    def get_tax(self):
        return 0
    def __str(self):
        return f'Book({super().__str__()})'

class Food(Product) :
    def get_tax(self):
        return self.get_price() * 0.03
    def __str(self):
        return f'Food({super().__str__()})'

#Products 리스트
#-요소로 Product를 상속받은 인스턴스만 저장
# Product 가 자식 타입인 TV, Book, Food로 인식되어
# Type Hint Error 가 발생하지 않는다.
# ==Product가 , TV, Book, Food 타입으로 변했다. ==
# -> 다형성
products : list[Product] = [
    tv1,
    tv2,
    Book('book-1','python for Beginners',20000),
    Food('apple-1','Apple',10000)
]

for prod in products :
    print('상품',prod)
    print('세금:',prod.get_tax())
```
