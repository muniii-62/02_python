## Getter/Setter 메서드
캡슐화된 클래스에서 속성을 제어하기 위한 메서드로써 기본적으로 다음 두 유형이 있다.
1. getter 속성의 값을 읽어내기 위한 메서드
2. setter 속성의 값을 설정하기 위한 메서드

```python
class MyClass :
    def __init__(self,value,name='123'):
        self.value = value
    #외부에서 private 속성 값을 얻어 가게 하는 메서드
    def get_value(self):
        return self.value

    #외부에서 private 속성 값을 변경(세팅)하는 메서드
    def set_value(self,value):
        self.value = value
```

```python
my_cls = MyClass(100)
print('my_cls.get_value():',my_cls.get_value())
print('200셋팅')
my_cls.set_value(200)
print('my_cls.get_value():',my_cls.get_value())
```

## Python의 `@property`

Python에서는 Java식 `get_value()`, `set_value()` 메서드도 사용할 수 있지만, 더 Python다운 방식으로는 `@property`를 자주 사용한다.

`@property`를 사용하면 메서드로 값을 검증하면서도, 사용하는 쪽에서는 속성처럼 읽고 쓸 수 있다.


```python
class Score :
    def __init__(self,value : int):
        self.value = value
    #오버로딩: 한 클래스 내에 같은 이름의 메서드가 여러 개 존재하는 기술
    @property
    def value(self): #get_value()
        return self.__value

    @value.setter
    def value(self,value):
        if not (0 <= value <= 100) :
            raise ValueError("점수는 0~100 사이만 세팅 가능")

        self.__value = value

score = Score(80)
print('score.value:',score.value)

score.value = 90
print("score.value:",score.value)

score.value = 120# 0~100
```
