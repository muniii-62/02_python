## 파일 쓰기

```python
f = open('helloworld2.txt','w',encoding='utf-8')
print(f,type(f))

text ="""
오늘 월요일
"""
f.write(text)
#기존 내용이 있으면 덮어쓰기
# 파일이 없으면 파일 생성

#파일 연결에 사용된 메모리 반환
f.close()
```

```python
# with 절
# with 옆에 작성된 open() 구문에서 사용한 파일 연결을
# with절 종료 시 자동으로 close()
with open('helloworld.txt','w',encoding='utf-8') as f:
    f.write("with 절을 이용하면 자동 close()된다")
```

## 파일 읽기

```python
with open('helloworld.txt','r',encoding='utf-8') as f:
    print(f.read())
```

```python
#한줄씩 읽기
with open('helloworld.txt','r',encoding='utf-8') as f:
    for i,line in enumerate(f):
        print(i+1,line,end='')
```

```python
# 파일 이어쓰기 ('a',append 모드 )
with open('helloworld.txt','a',encoding='utf-8') as f:
    f.write('''배고파
    home
    money
    ''')

with open('helloworld.txt','r',encoding='utf-8') as f:
    print(f.read())
```

```python
#파일 ori -> memory -> copy
#1. 파일 읽기
with open('helloworld.txt','r',encoding='utf-8') as f:
    text = f.read() #with절에 선언한 변수는 지역변수 x-> 전역 사요 가능


#2.파일 쓰기
with open('helloworld_copy.txt','w',encoding='utf-8') as f:
    f.write(text)
    print('파일 복사 완료')


```

```python
#이미지 파일 복사(아무 파일이나 가능)
# -r,w,a == 문자 읽기/쓰기/이어쓰기/ 모드
# -rb,wb ==바이트 단위 읽기/쓰기 모드
with open('skn33.png','rb') as f:
    #print(f.read())
    image_bytes = f.read()

with open('copy.png','wb') as f:
    f.write(image_bytes)
```

## dict-csv 읽고 쓰기

```python
#csv 읽기
#csv.DictReader(f)
#csv.DictWriter(f,fieldnames)
import csv
with open('users.csv','r',encoding='utf-8') as f:
    reader = csv.DictReader(f) # DictReader 객체 생성

    users = [row for row in reader]
    print(users)
```

```python
#csv 쓰기
with open('users.csv','w',encoding='utf-8', newline='') as f:

    #csv 쓰기 객체 생성
    #fl 쓸파일
    # -field :첫 줄에 작성될 헤더(컬럼명)
    writer = csv.DictWriter(f,fieldnames=('id','name','email'))
    writer.writeheader()# 필드네임으로 헤더를 작성

    #users에 사용자 추가
    users.append({'id':400,'name':'이순신','email':'lee@naver.com'})
    users.append({'id':500,'name':'강감찬','email':'gang@naver.com'})


    #여러 줄 쓰기
    # writer.writerows(users)

    for user in users :
        writer.writerow(user)
```
